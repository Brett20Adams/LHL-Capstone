## 1.2 - Assets, fuel type and regions EDA

```python
# How many unique assets are in this dataset?
generation_assets = df_generation['Asset Short Name'].value_counts().index.tolist()
print(len(generation_assets))
```

```python
# What are the different fuel type categories?
fuel_types = df_generation['Fuel Type'].value_counts().index.tolist()
fuel_types
```

```python
# How many different regions are there?
regions = df_generation['Region'].value_counts().index.tolist()
regions
```

#### There are six different regions that assets are classified into.

```python
# How many assets are in each region?
assets_per_region = df_generation.copy().drop_duplicates(subset=['Asset Short Name'])
assets_per_region['Region'].value_counts()
```

#### The majority of assets are in the South region. 

```python
# Let's check to see what the breakdown of fuel types in the South region is
south_region = assets_per_region[assets_per_region['Region'] == 'South']
south_region['Fuel Type'].value_counts()
```

#### Most energy assets in the South region are clean tech (Wind and Solar). Knowing that the oil sands area is in Northern Alberta, this makes sense, but let's confirm by repeating this check for the North regions.

```python
north_region = assets_per_region[(assets_per_region['Region'] == 'Northwest') | (assets_per_region['Region'] == 'Northeast')]
north_region['Fuel Type'].value_counts()
```

```python
# How many assets are in each fuel category?
assets_per_region['Fuel Type'].value_counts()
```

```python
# What is the average system capability of each fuel type?
(assets_per_region.groupby('Fuel Type')['System Capability'].sum())/(assets_per_region.groupby('Fuel Type')['Asset Name'].count())
```

```python
# What is the total capacity of each region?
assets_per_region.groupby('Region')['System Capability'].sum().sort_values(ascending=False)
```

#### The South region has the highest total system capability, which is correlated with having the most assets. Now, let's check to see what the average system capability is when adjusted for the number of assets in that region to determine the energy density of each region.

```python
(assets_per_region.groupby('Region')['System Capability'].sum()/(assets_per_region.groupby('Region')['Fuel Type'].count())).sort_values(ascending=False)
```

#### When we analyze the regions by energy density, however, the South region has the second lowest energy density, almost 3x lower than the Northeast region. This can be attributed to clean energy assets having lower energy densities than natural gas.

```python
# Total system capability volume for each fuel type
assets_per_region.groupby('Fuel Type')['System Capability'].sum().sort_values(ascending=False)
```

#### Create new dataframe that sums all the assets by fuel type

```python
df_price_3 = df_price.copy()
# add new fuel type totals
df_price_3['Total Coal Volume (MW)'] = df_price_3[coal_list].sum(axis=1)
df_price_3['Total Dual Fuel Volume (MW)'] = df_price_3[dual_fuel_list].sum(axis=1)
df_price_3['Total Energy Storage Volume (MW)'] = df_price_3[energy_storage_list].sum(axis=1)
df_price_3['Total Gas Volume (MW)'] = df_price_3[gas_list].sum(axis=1)
df_price_3['Total Hydro Volume (MW)'] = df_price_3[hydro_list].sum(axis=1)
df_price_3['Total Other Volume (MW)'] = df_price_3[other_list].sum(axis=1)
df_price_3['Total Solar Volume (MW)'] = df_price_3[solar_list].sum(axis=1)
df_price_3['Total Wind Volume (MW)'] = df_price_3[wind_list].sum(axis=1)
df_price_3['Total Volume (MW)'] = df_price_3[['Total Coal Volume (MW)', 'Total Dual Fuel Volume (MW)', 'Total Energy Storage Volume (MW)', 'Total Gas Volume (MW)',
            'Total Hydro Volume (MW)', 'Total Other Volume (MW)', 'Total Solar Volume (MW)', 'Total Wind Volume (MW)']].sum(axis=1)
# add new column summing up imports and export values (Net Actual Interchange)
df_price_3['Net Actual Interchange (MW)'] = df_price_3[['EXPORT_BC', 'EXPORT_MT', 'EXPORT_SK']].sum(axis=1) - df_price_3[['IMPORT_BC', 'IMPORT_MT', 'IMPORT_SK']].sum(axis=1)
```

```python
# remove individual asset columns
df_price_3a = df_price_3.iloc[:,:1]
df_price_3b = df_price_3.iloc[:,-19:]
df_price_3 = df_price_3a.join(df_price_3b)
# sum asset volumes to confirm calculated volume = total volume
df_price_3['Calculated Generated Volume'] = df_price[assets].sum(axis=1)
# reorganize columns
df_price_3 = df_price_3[['Date_Begin_Local', 'Total Coal Volume (MW)', 'Total Dual Fuel Volume (MW)', 'Total Energy Storage Volume (MW)', 'Total Gas Volume (MW)', 
'Total Hydro Volume (MW)', 'Total Other Volume (MW)', 'Total Solar Volume (MW)', 'Total Wind Volume (MW)', 'EXPORT_BC', 'EXPORT_MT', 'EXPORT_SK', 'IMPORT_BC', 
'IMPORT_MT', 'IMPORT_SK', 'Net Actual Interchange (MW)', 'Total Volume (MW)', 'Calculated Generated Volume', 'ACTUAL_AIL', 'ACTUAL_POOL_PRICE']]
df_price_3
```

```python
# convert date column to datetime
df_price_3['Date_Begin_Local'] = pd.to_datetime(df_price_3['Date_Begin_Local'])
```
